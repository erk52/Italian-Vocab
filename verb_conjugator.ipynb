```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
```

```python
def conjugate(infinitive):
    fil = infinitive + '.htm'
    with urllib.request.urlopen('http://italian.about.com/library/verb/blverb_'+fil) as response:
        html = response.read()
    soup = BeautifulSoup(html,'lxml')
    
    start = soup.get_text().find(infinitive+':') + len(infinitive)+2
    stop = soup.get_text()[start:].find('\n')
    definition = soup.get_text()[start:start+stop]
    
    lines = soup.get_text().split()
    s = lines.index('INDICATIVE/INDICATIVO')
    present_indicative = [lines[s+2][2:], lines[s+3][2:], lines[s+6][3:], lines[s+7][3:],
                          lines[s+8][3:], lines[s+10][4:]]
    imperfect_indicative = [lines[s+12][2:], lines[s+13][2:],lines[s+16][3:],lines[s+17][3:],
                            lines[s+18][3:],lines[s+20][4:]]
    simple_past = [lines[s+45][2:] + ' ' +lines[s+46], lines[s+47][2:] +' '+ lines[s+48], 
                   lines[s+51][3:] +' '+lines[s+52],lines[s+53][3:] +' '+lines[s+54], 
                   lines[s+55][3:] +' '+lines[s+56], lines[s+58][4:] +' '+lines[s+59]]
    abs_past = [lines[s+23][2:], lines[s+24][2:],lines[s+27][3:],lines[s+28][3:],
                lines[s+29][3:],lines[s+31][4:]]
    future= [lines[s+34][2:], lines[s+35][2:],lines[s+38][3:],lines[s+39][3:],
             lines[s+40][3:],lines[s+42][4:]]
    
    s = lines.index('SUBJUNCTIVE/CONGIUNTIVO')
    subj = [lines[s+2][2:],lines[s+3][2:],lines[s+6][3:],lines[s+7][3:],
            lines[s+8][3:],lines[s+10][4:]]
    past_subj = [lines[s+12][2:],lines[s+13][2:],lines[s+16][3:],lines[s+17][3:],
            lines[s+18][3:],lines[s+20][4:]]
    
    s = lines.index('CONDITIONAL/CONDIZIONALE')
    cond = [lines[s+2][2:],lines[s+3][2:],lines[s+6][3:],lines[s+7][3:],
            lines[s+8][3:],lines[s+10][4:]]
    
    return [definition,present_indicative,simple_past,imperfect_indicative, 
            abs_past,future,subj,past_subj,cond]
```

```python
def genTable(inf):
    pro = ['io', 'tu', 'lui/lei', 'noi','voi','loro']
    tense = ['Present Indicative','Passato Prossimo', 'Imperfetto', 'Passato Remoto', 'Futuro',
        'Present Subjunctive', 'Imperfect Subjunctive','Conditional']
    
    longt = []
    longp =[]
    for t in tense:
        longp += pro
        longt += [t for i in range(len(pro))]
    arrays = [longt, longp]

    v = conjugate(inf)
    dat = []
    for con in v[1:]:
        dat += con

    table = pd.DataFrame(index=arrays,data=dat).T
    table['Infinitive'] = inf
    table['Definition'] = v[0]
    
    return table
```

```python
def genVerbList():
    verbs = []
    for i in range(1,11):
        page = 'http://www.italian-verbs.com/italian-verbs/italian-verbs-top.php?pag='+str(i)
        with urllib.request.urlopen(page) as response:
            html = response.read()
        soup = BeautifulSoup(html,'lxml')
        lines = soup.get_text().split()
        for line in lines:
            if line.isalpha() and line.endswith('re'):
                verbs.append(line)
            
    return verbs
```

```python
def generateDatabase():
    verbs = genVerbList()
    MasterDF = pd.DataFrame()
    for v in verbs:
        try:
            temp = genTable(v)
            MasterDF = pd.concat([MasterDF,temp])
        except:
            print('Problem with '+v)
            continue
    return MasterDF
```

```python
VerbTable = generateDatabase()
```
